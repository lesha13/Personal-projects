```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
data = pd.read_csv(r"C:\Users\Льоша\video_dataset\USvideos.csv", parse_dates=True)
data.tail(5)
```

```python
all_vid = []

for _ in ["CA", "DE", "FR", "GB", "IN", "JP", "KR", "MX", "RU", "US"]:
    data = pd.read_csv(f"C:\\Users\\Льоша\\video_dataset\\{_}videos.csv")
    data["country"] = _
    all_vid.append(data)
```

```python
data = pd.concat(all_vid, ignore_index=True)
data
```

```python
data.info()
```

```python
data[["trending_date", "publish_time"]].tail(5) # strings
```

```python
data["publish_time"] = pd.to_datetime(data["publish_time"])
data["publish_time"].tail(5)
```

```python
data["trending_date"] = data["trending_date"].apply(lambda _: dt.strptime(_, "%y.%d.%m"))
data["trending_date"].tail(5)
```

```python
data["country"] = data["country"].map({
    "CA": "Canada",
    "DE": "Germany",
    "FR": "France:",
    "GB": "Great Britain", 
    "IN": "India", 
    "JP": "Japan",
    "KR": "South Korea",
    "MX": "Mexico",
    "RU": "Russia",
    "US": "USA"
})
data["country"].tail(5)
```

```python
# removed videos
data[data["video_error_or_removed"]].tail(5)
```

```python
# removing removed videos
data = data[~data["video_error_or_removed"]]
```

```python
# droping video_error_or_removed column cause we don't need it any more
data = data.drop(columns="video_error_or_removed")
```

```python
# creating new column
data.insert(11, "like_dislike_ratio", data.likes / (data.likes + data.dislikes))
```

```python
# creating new column
data.insert(12, "activity_per_viewer_ratio", (data.likes + data.dislikes + data.comment_count) / data.views)
```

```python
# average like/dislike ratio
data["like_dislike_ratio"].mean()
```

```python
# average like/dislike ratio per country
data.groupby("country")["like_dislike_ratio"].mean()
```

```python
data.groupby("country")["like_dislike_ratio"].mean().plot(kind="bar",
                                                          figsize=(10, 5),
                                                          rot=45)
```

```python
# average like/dislike ratio per category id
data.groupby("category_id")["like_dislike_ratio"].mean()
```

```python
data.groupby("category_id")["like_dislike_ratio"].mean().plot(kind='barh',
                                                              figsize=(10, 10))
```

```python
data.groupby(["country", "category_id"])["like_dislike_ratio"].mean()
```

```python
# average activity ratio
data["activity_per_viewer_ratio"].mean()
```

```python
# average activity ratio per country
data.groupby("country")["activity_per_viewer_ratio"].mean()
```

```python
data.groupby("country")["activity_per_viewer_ratio"].mean().plot(kind='bar',
                                                                 figsize=(10, 5), 
                                                                 rot=45)
```

```python
# average activity ratio per category id
data.groupby("category_id")["activity_per_viewer_ratio"].mean()
```

```python
data.groupby("category_id")["activity_per_viewer_ratio"].mean().plot(kind='barh',
                                                                     figsize=(10, 10))
```

```python
data.groupby(["country", "category_id"])["activity_per_viewer_ratio"].mean()
```
